```python
import csv
import pandas as pd
import matplotlib.pyplot as plt

time_line = []

pearl_buy = []
pearl_sell = []

banana_buy = []
banana_sell = []

with open('data_1.csv') as data:
    csv_reader = csv.reader(data, delimiter=';')
    for row in csv_reader:
        time_line.append(row[1])
        if row[2] == "PEARLS":
            pearl_buy.append(row[3])
            pearl_sell.append(row[9])
        elif row[2] == "BANANAS":
            banana_buy.append(row[3])
            banana_sell.append(row[9])


time_line = list(map(int, time_line[1:]))
pearl_buy = list(map(int, pearl_buy))
pearl_sell = list(map(int, pearl_sell))
banana_buy = list(map(int, banana_buy))
banana_sell = list(map(int, banana_sell))

time_stamp = []
for elem in time_line:
    if elem not in time_stamp:
        time_stamp.append(elem)
```

```python
print(pearl_buy)
print(time_stamp)

print(len(pearl_buy))
print(len(time_stamp))
```

```python
data = {'time_stamp': time_stamp,
        'pearl_buy': pearl_buy,
        'pearl_sell': pearl_sell}
df = pd.DataFrame(data)

df.plot(kind='line',x='time_stamp',linewidth=0.2)
plt.title('Pearl buy plot')
plt.xlabel('time_stamp')
plt.ylabel('pearl_buy')
plt.show()
```

```python
data = {'time_stamp': time_stamp,
        'banana_buy': banana_buy,
        'banana_sell': banana_sell}
df = pd.DataFrame(data)

df.plot(kind='line',x='time_stamp',linewidth=0.2)
plt.title('banana_buy plot')
plt.xlabel('time_stamp')
plt.ylabel('banana_buy')
plt.show()
```
